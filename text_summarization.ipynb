### Text Summarization

```python
text = '''
Multi-Document Text Summarization (MDTS) consists of generating 
an abstract from a group of two or more number of documents that represent only 
the most important information of all documents. Generally, the objective is to 
obtain the main idea of several documents on the same topic. In this paper, we 
propose a new MDTS method based on a genetic algorithm (GA). The fitness 
function is calculated considerer two text features: sentence position and coverage. We propose the binary coding representation, selection, crossover and mutation operators to improve the state-of-the-art results. We test the proposed 
method on DUC02 data set specifically on Abstractive Multi-Document Text 
Summarization (AMDST) task demonstrating the improvement over the state-ofart methods. Four different tasks for each of the 59 collection of documents are 
tested. In addition, we test different configurations of the most used methodology 
to generate AMDST summaries. Moreover, different heuristics such as topline, 
baseline, baseline-random and lead baseline are calculated. The proposed method 
for AMDTS demonstrates the improvement over the state-of-art methods and 
heuristics.
'''
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stop_word = list(STOP_WORDS)
stop_word
```

```python
nlp = spacy.load('en_core_web_sm')
docs = nlp(text)
```

```python
docs
```

```python
token = [token.text for token in docs]
token
```

```python
punctuation = punctuation + '\n'
punctuation
```

```python
word_frequency = {}
for word in docs:
    if word.text.lower() not in punctuation:
        if word.text.lower() not in stop_word:
            if word.text not in word_frequency.keys():
                word_frequency[word.text]=1
            else:
                word_frequency[word.text] +=1
```

```python
word_frequency
```

```python
max_frequencies = max(word_frequency.values())
```

```python
max_frequencies
```

```python
for word in word_frequency.keys():
    word_frequency[word] = word_frequency[word]/max_frequencies
```

```python
print(word_frequency)
```

```python
sent_token = [sent for sent in docs.sents]
print(sent_token)
```

```python
sentence_score = {}
for sent in sent_token:
    for word in sent:
        if word.text.lower() in word_frequency.keys():
            if sent not in sentence_score.keys():
                sentence_score[sent] = word_frequency[word.text.lower()]
            else:
                sentence_score[sent] += word_frequency[word.text.lower()]
```

```python
sentence_score
```

```python
sentence_length =  int(len(sent_token)*0.3)
sentence_length
```

```python
from heapq import nlargest
summary = nlargest(sentence_length, sentence_score, key = sentence_score.get)
```

```python
summary
```

```python
final_summary = [word.text for word in summary]
final_summary
```

```python
summary = ' '.join(final_summary)
summary
```
